```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('D:/DataSets/insurance.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df_southwest=df.loc[df['region']=='southwest']
```

```python
df_northwest=df.loc[df['region']=='northwest']
df_southeast=df.loc[df['region']=='southeast']
```

```python
plt.plot(df_southwest['region'],np.zeros_like(df_southwest['region']),'o')
plt.plot(df_northwest['region'],np.zeros_like(df_northwest['region']),'o')
plt.plot(df_southeast['region'],np.zeros_like(df_southeast['region']),'o')
plt.xlabel('REGION')
plt.show()
```

```python
sns.FacetGrid(df,hue="region",size=5).map(plt.scatter,"bmi","charges").add_legend();
plt.show()
```
